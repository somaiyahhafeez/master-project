```python
raw_headers = df.iloc[0].tolist()

cleaned_headers = raw_headers[:2] + [
    col.replace("Estimate!!Total:!!", "")
       .replace("Estimate!!Total:", "")
       .replace("!!", ":")
       .strip(":")
    for col in raw_headers[2:]
]

df.columns = cleaned_headers
df = df[1:].reset_index(drop=True)
df
```
