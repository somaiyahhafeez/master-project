```python
cutoff_date = New_York["Arrest Date"].max()

ytd_2025 = (New_York["Arrest Date"].dt.year == 2025) & \
           (New_York["Arrest Date"].dt.month <= cutoff_date.month) & \
           (New_York["Arrest Date"].dt.day <= cutoff_date.day)

ytd_2024 = (New_York["Arrest Date"].dt.year == 2024) & \
           (New_York["Arrest Date"].dt.month <= cutoff_date.month) & \
           (New_York["Arrest Date"].dt.day <= cutoff_date.day)

df_ytd = New_York[ytd_2024 | ytd_2025]
df_ytd["Year"] = df_ytd["Arrest Date"].dt.year

countries_ytd = (
    df_ytd.groupby(["Citizenship Country", "Year"])
          .size()
          .reset_index(name="Arrests")
)

pivot_ytd = countries_ytd.pivot(index="Citizenship Country", columns="Year", values="Arrests").fillna(0)

for col in [2024, 2025]:
    if col not in pivot_ytd.columns:
        pivot_ytd[col] = 0

pivot_ytd["% of Total Arrests"] = (pivot_ytd[2025] / pivot_ytd[2025].sum()) * 100

def custom_change(row):
    if row[2024] == 0:
        return f"+{int(row[2025])} arrests"
    else:
        change = ((row[2025] - row[2024]) / row[2024]) * 100
        sign = "+" if change > 0 else ""
        return f"{sign}{change:.0f}%"

pivot_ytd["Change from 2024"] = pivot_ytd.apply(custom_change, axis=1)

final_ytd = pivot_ytd.reset_index().sort_values(by=2025, ascending=False)
final_ytd
```

```python
final_ytd.to_csv("arrests_per_country.csv")
```

## age of those arrested

```python
New_York["Age"] = 2025 - New_York["Birth Year"]

filtered = New_York[New_York["year"].isin([2024, 2025])]

age_counts_by_year = (
    filtered.groupby(["year", "Age"])
    .size()
    .reset_index(name="Count")
    .sort_values(["year", "Age"])
)


age_stats_by_year = (
    filtered.groupby("year")["Age"]
    .agg(
        Min_Age="min",
        Max_Age="max",
        Mean_Age="mean"
    )
    .reset_index()
)


def min_max_counts(group):
    min_age = group["Age"].min()
    max_age = group["Age"].max()
    median_age = group["Age"].median()

    min_count = (group["Age"] == min_age).sum()
    max_count = (group["Age"] == max_age).sum()

    return pd.Series({
        "Min_Age": min_age,
        "Min_Age_Count": min_count,
        "Max_Age": max_age,
        "Max_Age_Count": max_count,
        "Median_Age": median_age
    })

age_stats_by_year = filtered.groupby("year").apply(min_max_counts).reset_index()


case_status_per_year = (
    New_York[New_York["year"].isin([2024, 2025])]
    .groupby(["year", "Case Category"])
    .size()
    .reset_index(name="Count")
)

print(case_status_per_year)



```
