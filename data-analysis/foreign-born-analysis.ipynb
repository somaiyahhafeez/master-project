# loading data

```python
import pandas as pd
df=pd.read_csv("B05006-original-data.csv")
df
```

# truncating GEOID to match with the dataframe to merge on

```python
print(df.dtypes)
```

```python
df['GEO_ID'] = df['GEO_ID'].str[9:]
```

```python
df
```

# reading file to merge on

```python
nta=pd.read_csv("Census_to_NTA_2020.csv")
nta
```

```python
nta = nta.rename(columns={'GEOID': 'GEO_ID'})
```

# merging

```python
df.columns = df.iloc[0]  

df

```

```python
df = df.rename(columns={'': 'GEO_ID'})
df
```

```python
df['GEO_ID'] = df['GEO_ID'].astype(str)

nta['GEO_ID'] = nta['GEO_ID'].astype(str)

merged_df = pd.merge(df, nta, on='GEO_ID')

```

```python
merged_df
```

```python
merged_df.columns

```

# removing margin of errors and other unwanted columns

```python
merged_df.columns=merged_df.columns.astype(str)

```

```python
merged_df = merged_df.loc[:, ~merged_df.columns.str.contains("Margin of Error!!", regex=False)]
merged_df
```

```python
merged_df = merged_df.drop(columns=['nan', 'CountyFIPS', 'BoroCode','BoroCT2020', 'CT2020', 'CTLabel', 'NTACode', 'NTAType', 'NTAAbbrev', 'CDTACode', 'CDTAType', 'CDTAName'])
merged_df
```

# cleaning header

```python
merged_df.columns = (
    list(merged_df.columns[:3]) + [
        str(col)
        .replace("Estimate!!Total:!!", "")
        .replace("Estimate!!Total:", "")
        .replace("!!", ":")
        .strip(":")
        for col in merged_df.columns[3:]
    ]
)
```

```python
merged_df
```

# selecting just subregions for now

```python
selected_cols = [
    'GEO_ID',
    'Geographic Area Name',
    'Estimate!!Total:',
    'Europe::Northern Europe',
    'Europe::Western Europe',
    'Europe::Southern Europe',
    'Europe::Eastern Europe',
    'Asia::Eastern Asia',
    'Asia::South Central Asia',
    'Asia::South Eastern Asia',
    'Asia::Western Asia',
    'Africa::Eastern Africa',
    'Africa::Middle Africa',
    'Africa::Northern Africa',
    'Africa::Southern Africa',
    'Africa::Western Africa',
    'Oceania',
    'Americas::Latin America',
    'Americas::Latin America::Central America',
    'Americas::Latin America::South America',
    'Americas::Northern America',
    'BoroName',
    'NTAName'
]

region_df = merged_df[selected_cols].copy()
region_df
```

# merging with total population from table B01003

```python
total_pop=pd.read_csv("total_population.csv")
total_pop
```

```python
total_pop = total_pop.rename(columns={'NAME': 'Geographic Area Name'})
total_pop
```

```python
total_pop = total_pop.drop(columns=["B01003_001M", "Unnamed: 4"])
```

```python
total_pop['GEO_ID'] = total_pop['GEO_ID'].str[9:]
```

```python
merged_df_2 = pd.merge(region_df, total_pop, on='GEO_ID', how='inner')  
```

```python
merged_df_2 = merged_df_2.rename(columns={
    'Estimate!!Total:': 'total_foreign_born',
    'B01003_001E': 'total_tract_pop'
})
merged_df_2
```

```python
merged_df_2["total_foreign_born"] = pd.to_numeric(merged_df_2["total_foreign_born"], errors='coerce')
merged_df_2["total_tract_pop"] = pd.to_numeric(merged_df_2["total_tract_pop"], errors='coerce')
```

```python
print(merged_df_2.dtypes)
```

# grouping by neighborhood

```python
columns_to_convert = [
    'Europe::Northern Europe',
    'Europe::Western Europe',
    'Europe::Southern Europe',
    'Europe::Eastern Europe',
    'Asia::Eastern Asia',
    'Asia::South Central Asia',
    'Asia::South Eastern Asia',
    'Asia::Western Asia',
    'Africa::Eastern Africa',
    'Africa::Middle Africa',
    'Africa::Northern Africa',
    'Africa::Southern Africa',
    'Africa::Western Africa',
    'Oceania',
    'Americas::Latin America',
    'Americas::Latin America::Central America',
    'Americas::Latin America::South America',
    'Americas::Northern America',    
]

for col in columns_to_convert:
    merged_df_2[col] = pd.to_numeric(merged_df_2[col], errors="coerce")
```

```python
nta_summary = merged_df_2.groupby("NTAName", as_index=False).sum(numeric_only=True)
```

```python
nta_summary
```

# calculating foreign born percentage population 

```python
nta_summary = nta_summary.dropna()
nta_summary
```

```python
nta_summary["foreign_born_pct"] = (nta_summary["total_foreign_born"] / nta_summary["total_tract_pop"]) * 100
nta_summary.sort_values(by=["foreign_born_pct"], ascending=False).head(50)
```

```python
for col in nta_summary.select_dtypes(include='number').columns:
    if col != "total_tract_pop":
        nta_summary[f"{col}_pct"] = (nta_summary[col] / nta_summary["total_tract_pop"]) * 100
```

```python
nta_summary.columns
```

```python
nta_summary.sort_values(by="Asia::South Central Asia_pct", ascending=False).head(50)
```
